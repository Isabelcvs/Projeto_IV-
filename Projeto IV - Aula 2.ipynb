# **Referencial Teórico**

<div align="justify"> <p> A análise de séries temporais constitui uma das abordagens mais relevantes para investigar fenômenos socioeconômicos que apresentam comportamento periódico ou evolutivo, como a dinâmica do emprego formal no Brasil. De acordo com Morettin e Toloi (2018), uma série temporal é formada por observações coletadas em intervalos regulares, nas quais cada valor mantém correlação com os anteriores, permitindo a identificação de padrões de tendência, sazonalidade, ciclos e ruído. Esse tipo de análise não apenas descreve eventos passados, mas também possibilita a elaboração de previsões e cenários prospectivos, o que se mostra essencial para a formulação de políticas públicas e estratégias organizacionais.

Entre as metodologias clássicas, destacam-se os modelos ARIMA (AutoRegressive Integrated Moving Average), popularizados por Box, Jenkins e Reinsel (2015), amplamente aplicados em contextos de séries estacionárias. A principal vantagem desses modelos é a flexibilidade para captar relações autorregressivas e médias móveis; contudo, sua limitação reside na necessidade de estacionariedade e na complexidade de parametrização, o que pode dificultar sua aplicação em séries mais curtas ou com elevada variabilidade.

Outra linha de abordagem é a dos métodos de suavização exponencial, como Holt-Winters, amplamente reconhecidos pela eficiência em capturar tendências e sazonalidades em dados econômicos (HYNDMAN; ATHANASOPOULOS, 2018). Tais métodos apresentam simplicidade operacional e bom desempenho em horizontes curtos, ainda que sejam sensíveis a choques abruptos. Em complemento, técnicas como médias móveis — simples, ponderadas ou exponenciais — oferecem recursos interpretativos valiosos para destacar tendências e reduzir ruídos, ainda que apresentem limitações na modelagem de relações complexas com variáveis externas.

As técnicas de decomposição clássica, por sua vez, permitem separar a série em seus componentes estruturais — tendência, sazonalidade e resíduo — favorecendo análises descritivas, comparações setoriais e diagnósticos regionais (CLEVELAND et al., 1990). Embora não sejam voltadas prioritariamente para previsão, constituem recurso importante para a caracterização do fenômeno em estudos exploratórios.

Mais recentemente, avanços no campo da Ciência de Dados têm possibilitado o desenvolvimento de modelos híbridos, que combinam métodos estatísticos tradicionais com técnicas de aprendizado profundo. Pesquisas publicadas no IEEE Xplore indicam que a integração de modelos como ARIMA com redes neurais recorrentes (LSTM e GRU) proporciona ganhos significativos de acurácia, sobretudo em séries não lineares ou com padrões de alta variabilidade (ZHANG, 2003; LI et al., 2021). Esses modelos apresentam como vantagens a flexibilidade e a capacidade de capturar dinâmicas complexas; entretanto, demandam maior volume de dados, infraestrutura computacional e maior complexidade interpretativa.

No caso específico do mercado de trabalho formal brasileiro, o Novo CAGED disponibiliza mensalmente informações sobre admissões, desligamentos, saldo e estoque de vínculos celetistas, constituindo fonte primária para análises de políticas de emprego (BRASIL, 2025). O Sumário Executivo funciona como régua de validação, apresentando sínteses nacionais, recortes setoriais, regionais e de perfil de trabalhadores (sexo, idade, escolaridade, faixas salariais). Embora esses relatórios assegurem confiabilidade e padronização, apresentam limitações de granularidade quando comparados aos microdados completos.

Diante disso, a solução proposta neste trabalho busca aliar a clareza interpretativa dos métodos estatísticos clássicos à capacidade descritiva da decomposição e da análise exploratória de dados. O pipeline proposto contempla tanto a caracterização dos saldos por setor, região e perfil quanto a proposição de indicadores sintéticos, de forma a gerar relatórios replicáveis e de fácil interpretação para gestores públicos e privados. Dessa forma, o projeto se ancora em uma tradição metodológica robusta e, ao mesmo tempo, mantém abertura para incorporar técnicas mais sofisticadas em fases posteriores, em consonância com práticas recentes discutidas na literatura internacional.


# **Pipeline da Solução**

Essa organização metodológica encontra respaldo em propostas contemporâneas de data pipelines voltados para análises socioeconômicas, que ressaltam a importância de fluxos padronizados, escaláveis e auditáveis (PENG, 2020; LI et al., 2021).

### **1. Coleta de dados**

A etapa de coleta de dados será responsável por reunir e organizar as bases oficiais que servirão de insumo para todas as análises subsequentes. Serão utilizadas duas fontes principais disponibilizadas pelo Ministério do Trabalho e Emprego (MTE): o Sumário Executivo do Novo CAGED (PDF), que trará os resultados consolidados em nível nacional, setorial, regional e de perfil dos trabalhadores; e a planilha “3-tabelas_Junho de 2025” (Excel), que reunirá informações em maior nível de detalhamento, abrangendo séries históricas desde 2020, além de desagregações por unidades federativas, municípios e setores de atividade econômica.

A caracterização dessas bases permitirá compreender a natureza e a abrangência dos dados. O Sumário Executivo será utilizado como régua oficial de validação, fornecendo totais consolidados e indicadores de perfil (sexo, idade, escolaridade e faixa salarial). Já a planilha em Excel representará a base operacional do pipeline, contendo microdados agregados que possibilitarão comparações entre regiões, setores e faixas de tempo. Dessa forma, será possível conciliar consistência institucional com granularidade analítica.

Além da caracterização das fontes, a etapa de coleta será planejada de modo a garantir que os principais componentes de uma série temporal estejam preservados e possam ser identificados no processo analítico. A tendência será assegurada pela extensão da série histórica (2020–2025), permitindo observar movimentos de longo prazo, como a recuperação do mercado de trabalho após a pandemia. A sazonalidade será garantida pela periodicidade mensal contínua dos dados, essencial para identificar padrões recorrentes em determinados meses, como contratações no comércio no final do ano. Os ciclos econômicos, por sua vez, serão considerados a partir das séries setoriais e regionais, possibilitando a detecção de flutuações associadas a fases de expansão ou retração da economia. Por fim, o ruído será preservado por meio da coleta dos valores brutos, incluindo registros classificados como “não informados” ou oscilações abruptas, que posteriormente serão tratados nas etapas de pré-processamento e suavização.

Assim, a coleta de dados não terá apenas a função de reunir os insumos, mas também de estruturá-los de maneira a permitir que cada componente fundamental da série temporal seja identificado e mensurado. Essa organização inicial será decisiva para a qualidade das análises subsequentes, pois garantirá que tanto a consistência institucional quanto a riqueza analítica sejam preservadas desde o início do pipeline.


### **2. Processamento dos Dados (Pré-Processamento)**
Na etapa de pré-processamento, os dados coletados passarão por um conjunto estruturado de atividades que terão como objetivo garantir consistência, qualidade e adequação ao formato exigido pela análise de séries temporais. As principais atividades serão:

**Normalização de rótulos e cabeçalhos**: os nomes das variáveis e categorias serão padronizados por meio de técnicas de string normalization, como remoção de acentos, unificação de maiúsculas e minúsculas e substituição de espaços por underscores. Essa prática garantirá uniformidade e facilitará a manipulação computacional dos dados (MCKINNEY, 2017).

**Conversão e tipagem de variáveis**: as variáveis numéricas, como admissões, desligamentos, saldo e estoque, serão convertidas para tipos numéricos adequados (int64 ou float64), enquanto as variáveis de tempo serão transformadas em objetos datetime. Isso permitirá precisão nos cálculos e padronização da série temporal (MORETTIN; TOLOI, 2018).

**Construção da chave temporal**: será criada a variável periodo_ref, padronizada no formato “YYYY-MM”, que servirá como índice temporal fixo para toda a série. Essa chave assegurará que os dados sejam tratados em frequência mensal, condição essencial para análises de tendência e sazonalidade (SHUMWAY; STOFFER, 2017).

**Organização das dimensões geográficas e setoriais**: as informações de Brasil, regiões, unidades federativas e municípios serão harmonizadas conforme o padrão do IBGE, enquanto os setores de atividade serão alinhados à classificação CNAE. Essa etapa garantirá comparabilidade entre diferentes recortes e evitará duplicidades (PENG, 2020).

**Tratamento de valores faltantes**: valores ausentes serão tratados de acordo com a natureza da variável. Para séries agregadas, será aplicada a técnica de imputação por valores anteriores (forward fill), assegurando continuidade. Para variáveis categóricas, como escolaridade ou faixa salarial, será utilizada a categoria “não informado”, preservando a integridade do banco de dados (HYNDMAN; ATHANASOPOULOS, 2018).

**Checagem de consistência aritmética**: será verificado se o saldo corresponde à diferença entre admissões e desligamentos em todos os registros. Eventuais inconsistências serão corrigidas com base nas informações mais granulares disponíveis, reforçando a confiabilidade da base de dados (BRASIL, 2025).

**Criação de indicadores derivados**: sSerão gerados indicadores adicionais, como o saldo por 10 mil vínculos ativos, que permitirá comparações mais justas entre regiões de diferentes portes. Também serão construídos acumulados de 12 meses, prática recomendada para suavizar variações de curto prazo em séries econômicas (CLEVELAND et al., 1990).

**Separação de estoque com e sem ajuste**: as duas versões do estoque fornecidas pelo Novo CAGED — “com ajuste” e “sem ajuste” — serão mantidas, possibilitando análises comparativas e validação cruzada entre as metodologias de apuração. Essa prática aumentará a transparência e a robustez dos resultados.


### **3. Análise exploratória e visualização**
A análise exploratória (EDA) será conduzida com o objetivo de descrever padrões e revelar insights preliminares. Serão produzidas séries temporais agregadas para o saldo nacional (2020–2025), rankings de saldos por setor, UF e município, além de mapas coropléticos e heatmaps que evidenciarão a intensidade relativa da geração de empregos.

Serão também geradas visualizações para caracterizar o perfil dos vínculos (sexo, idade, escolaridade e faixa salarial), de acordo com os dados disponíveis no Sumário Executivo. Complementarmente, serão aplicadas técnicas de médias móveis e funções de autocorrelação (ACF e PACF), conforme recomendadas pela literatura, para identificar dependências temporais e padrões estruturais (SHUMWAY; STOFFER, 2017; HYNDMAN; ATHANASOPOULOS, 2018).


### **4. Modelagem**
A etapa de modelagem será conduzida com base em métodos estatísticos clássicos e de interpretação transparente, priorizando técnicas capazes de atender aos objetivos do projeto sem a necessidade de ferramentas preditivas excessivamente complexas. Inicialmente, serão aplicados os modelos Naive e Sazonal Naive, que projetarão os valores futuros de forma simples, utilizando como referência o último valor observado ou o valor do mesmo mês em anos anteriores. Apesar da simplicidade, esses modelos servirão como linha de base (benchmarks), permitindo avaliar a efetividade de abordagens mais elaboradas.

Em seguida, será realizada a decomposição clássica da série temporal, que permitirá separar os dados em três componentes principais: tendência, sazonalidade e resíduo. A análise desses elementos possibilitará identificar, por exemplo, se existe uma trajetória de crescimento sustentado do emprego formal após 2020, se determinados meses apresentam sazonalidade recorrente (como contratações no comércio em dezembro) e se picos isolados podem ser classificados como ruído.

Complementarmente, será aplicada a técnica de suavização exponencial de Holt-Winters, que incorporará parâmetros para capturar tendência e sazonalidade de forma adaptativa. Esse método permitirá elaborar previsões de curto prazo, como a estimativa de como os saldos poderão se comportar nos meses subsequentes a junho/2025, com base nos padrões históricos da série.

Por fim, será considerada a utilização de modelos ARIMA (AutoRegressive Integrated Moving Average), que combinam componentes autorregressivos, diferenciação e médias móveis. A aplicação dependerá da verificação da estacionariedade das séries, avaliada por meio de testes estatísticos específicos, e será realizada de forma exploratória em séries que apresentarem forte dependência temporal, como as do setor de serviços. Ainda que o foco do projeto não esteja na previsão de longo prazo, os modelos ARIMA contribuirão para consolidar análises de séries temporais curtas e médias.

Embora o escopo principal seja a utilização de métodos descritivos e de média complexidade, reconhece-se que abordagens híbridas, como a combinação de ARIMA com redes neurais recorrentes (LSTM e GRU), vêm sendo exploradas em estudos recentes indexados no IEEE Xplore. Contudo, tais técnicas não serão implementadas neste trabalho, sendo apenas consideradas como possibilidade futura em pesquisas posteriores.

### **5. Avaliação e Validação**

A etapa de avaliação e validação será conduzida em duas dimensões complementares. Na dimensão estatística, serão aplicadas métricas de erro como o Erro Absoluto Médio (MAE), a Raiz do Erro Quadrático Médio (RMSE) e o Erro Percentual Absoluto Médio (MAPE), que permitirão mensurar a proximidade entre os valores ajustados pelos modelos e os dados efetivamente observados. Adicionalmente, serão realizados testes de estacionariedade, como o Augmented Dickey-Fuller (ADF), que indicarão a adequação das séries ao uso de modelos ARIMA, além da análise da força da tendência e da sazonalidade, possibilitando concluir, por exemplo, se o setor de comércio apresenta sazonalidade mais forte que o setor industrial.

Já na dimensão institucional, a validação ocorrerá por meio da comparação dos resultados produzidos pelo pipeline com os totais oficiais divulgados no Sumário Executivo do Novo CAGED. Esse processo garantirá que os saldos, admissões e desligamentos processados coincidam com os números oficiais reportados pelo Ministério do Trabalho e Emprego. Dessa forma, será assegurada não apenas a consistência estatística dos modelos aplicados, mas também a conformidade das análises com a régua institucional de referência.

Essa abordagem integrada permitirá validar o pipeline em termos de qualidade metodológica e relevância prática, fornecendo análises reprodutíveis e úteis para gestores públicos e privados compreenderem a movimentação do emprego formal no Brasil em junho de 2025.
